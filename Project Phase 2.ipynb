# Data Cleaning
Handle Missing Values


```python
# Check for missing values in zomato_data
missing_values = zomato_data.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

```python
# Check for duplicates
zomato_data.duplicated().sum()
```

```python
#  Remove duplicates if any
zomato_data = zomato_data.drop_duplicates()
```

```python
# Summary statistics for numerical columns
print(zomato_data.describe())
```

```python
from sklearn.impute import SimpleImputer

# Identify numerical columns
num_cols = ['Delivery_person_Age', 'Delivery_person_Ratings', 'multiple_deliveries']  # Add other numerical columns as needed

# Impute missing numerical values with the mean
num_imputer = SimpleImputer(strategy='mean')
zomato_data[num_cols] = num_imputer.fit_transform(zomato_data[num_cols])
```

```python
# Identify categorical columns
cat_cols = ['Weather_conditions', 'Road_traffic_density', 'Festival', 'City']  # Add other categorical columns as needed

# Impute missing categorical values with the most frequent value (mode)
cat_imputer = SimpleImputer(strategy='most_frequent')
zomato_data[cat_cols] = cat_imputer.fit_transform(zomato_data[cat_cols])
```

```python
# Replace null values with the most frequent value (mode)
most_frequent_time = zomato_data['Time_Orderd'].mode()[0]
zomato_data['Time_Orderd'] = zomato_data['Time_Orderd'].fillna(most_frequent_time)

# Check the dataset after replacing null values
print(zomato_data.head(10))
```

```python
# Check again to ensure no missing values remain
zomato_data.isnull().sum()
```

```python
# 3. Ensure correct data types
# Convert 'Order_Date' to datetime
zomato_data['Order_Date'] = pd.to_datetime(zomato_data['Order_Date'], errors='coerce')
```

```python
# Fill missing dates with the median date
median_date = zomato_data['Order_Date'].median()
zomato_data['Order_Date'].fillna(median_date, inplace=True)
```

```python
# Converting numerical columns to appropriate types
numerical_columns = ['Delivery_person_Age', 'Delivery_person_Ratings', 'Restaurant_latitude', 'Restaurant_longitude',
                     'Delivery_location_latitude', 'Delivery_location_longitude', 'Vehicle_condition', 'multiple_deliveries', 'Time_taken (min)']
for column in numerical_columns:
    zomato_data[column] = pd.to_numeric(zomato_data[column], errors='coerce')
    
zomato_data.info()
```

```python
# Check for rows with invalid time formats in both columns
invalid_time_orderd = zomato_data[~zomato_data['Time_Orderd'].str.match(r'^\d{2}:\d{2}$')]
invalid_time_order_picked = zomato_data[~zomato_data['Time_Order_picked'].str.match(r'^\d{2}:\d{2}$')]

# Drop rows with invalid time formats (if necessary)
zomato_data = zomato_data[zomato_data['Time_Orderd'].str.match(r'^\d{2}:\d{2}$')]
zomato_data = zomato_data[zomato_data['Time_Order_picked'].str.match(r'^\d{2}:\d{2}$')]
```

```python
# Convert other time columns ('Time_Orderd', 'Time_Order_picked') to datetime
zomato_data['Time_Orderd'] = pd.to_datetime(zomato_data['Time_Orderd'], format='%H:%M', errors='coerce')
zomato_data['Time_Order_picked'] = pd.to_datetime(zomato_data['Time_Order_picked'], format='%H:%M', errors='coerce')
```

```python
# Convert 'Time_taken (min)' to numeric
zomato_data['Time_taken (min)'] = pd.to_numeric(zomato_data['Time_taken (min)'], errors='coerce')
```

```python
# Convert time columns to datetime objects
zomato_data['Time_Orderd'] = pd.to_datetime(zomato_data['Time_Orderd'], format='%H:%M')
zomato_data['Time_Order_picked'] = pd.to_datetime(zomato_data['Time_Order_picked'], format='%H:%M')

# Calculate time difference in minutes
zomato_data['Order_Time'] = (zomato_data['Time_Order_picked'] - zomato_data['Time_Orderd']).dt.total_seconds() / 60
```

```python

zomato_data['Order_Date'] = pd.to_datetime(zomato_data['Order_Date'], format='%d-%m-%Y', errors='coerce')

# Fix for Time_Ordered
def convert_time_format(time_str):
    try:
        return pd.to_datetime(time_str, format='%H:%M').time()
    except ValueError:
        return np.nan

zomato_data['Time_Orderd'] = zomato_data['Time_Orderd'].apply(convert_time_format)
zomato_data['Time_Order_picked'] = zomato_data['Time_Order_picked'].apply(convert_time_format)
```

```python
print(zomato_data.head(10))
```

```python
# Drop unnecessary columns
zomato_data.drop(columns=['ID', 'Delivery_person_ID'], inplace=True)
```

```python
# Extract features from datetime
zomato_data['Order_Day'] = zomato_data['Order_Date'].dt.dayofweek  # Day of the week

# Convert time to datetime to extract hour
zomato_data['Order_Hour'] = zomato_data['Time_Orderd'].apply(lambda x: x.hour)

# Define peak hours
def time_slot(hour):
    if 8 <= hour < 10:
        return 'Breakfast'
    elif 12 <= hour < 14:
        return 'Lunch'
    elif 19 <= hour < 23:
        return 'Dinner'
    else:
        return 'Non-Peak'

zomato_data['Time_Slot'] = zomato_data['Order_Hour'].apply(time_slot)
```

# City with highest average delivery person rating

# Most Common food and Least Common food ordered

```python
# Get the counts of each order type
order_counts = zomato_data['Type_of_order'].value_counts()

# Find the most common order type and its count
most_common_order_type = order_counts.idxmax()
most_common_order_count = order_counts.max()

print(f"The most common type of order is '{most_common_order_type}', ordered {most_common_order_count} times.")
```

```python
# Get the counts of each order type
order_counts = zomato_data['Type_of_order'].value_counts()

# Find the least common order type and its count
least_common_order_type = order_counts.idxmin()
least_common_order_count = order_counts.min()

print(f"The least common type of order is '{least_common_order_type}', ordered {least_common_order_count} times.")

```
